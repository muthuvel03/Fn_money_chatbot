# **Sub chain - to handle chat history**

```python
import json
# Define the sub_chain function as provided
def sub_chain(query, chat_history):
    sub_chain = ChatPromptTemplate.from_messages([
        ("system", """
                You have two tasks to perform based on the user's question.

                **Task 1: Reformulate the Question**
                You are tasked with reformulating a user question into a standalone version that is clear and does not rely on previous chat history or references.

                Instructions for Task 1:
                1. Reformulate the question so that it can be understood independently, without needing prior chat history.
                2. Ignore any phrases like "last question," "as mentioned," "what I asked before," etc.
                3. If the question is already standalone or doesn't need reformulation (e.g., greetings, conclusions, irrelevant queries), return it as is.
                4. Focus on producing a concise and standalone question.

                **Examples for Task 1:**
                - Input: "What was my last question?"
                  Reformulated: "Can you explain how Bitcoin works?" // since the question refers to chat history

                **Task 2: Determine API Requirement**
                After reformulating the question, determine if the reformulated question pertains to current market insights.

                Instructions for Task 2:
                1. If the reformulated question relates to current market insights (e.g., "What’s happening in the market today?") which require today data (eg., uptodate today data), indicate this in the output with a key-value pair of "api_required": "yes".
                2. If it does not relate to current market insights, set "api_required" to "no".
                3. give 'yes' if only it requires today data regardless of it needs historical data or not.
                4. If it needs historical data then give 'no', elif it needs today data then give 'yes'.

                **Output Structure:**
                Return a JSON structure:
                  {{
                    "reformulated_question": reformulated question",
                    "api_required": "yes or no", // yes for current information or else no
                    "reason": "explanation for api_required result" // does it need today live info
                    }}


                Just give me the data inside the brackets , Strictly Do not include json string, the final response must be a type of JSON.


        """),

        ("human", """Chat History:\n{chat_history}\nUser Question:\n{question}"""),
    ]) | llm | StrOutputParser()


    # Invoke the sub-chain to process the query
    c_query = sub_chain.invoke({
        "chat_history": chat_history,
        "question": query
    })

    return c_query

# Function to invoke the chain with user query and initial empty chat history
def invoke_chain(user_query):
    # Load the chat history, which will be empty initially
    chat_history = memory.load_memory_variables({}).get("chat_history", "")
    print(chat_history)

    # Call the sub_chain function with the user query and the chat history
    reformulated_query = sub_chain(user_query, chat_history)


    return reformulated_query
# Example user query to test
user_query = "Give me the data for today market trend"

# Invoke the chain with the user query and empty chat history
final_question = invoke_chain(user_query)
print(final_question)
final_question1 = json.loads(final_question)
final_question = final_question1['reformulated_question']
api_req = final_question1['api_required']

```

# ***Live data collection***

```python
import requests

def fetch_live_data(symbol):
    api_key = "SQWD6PM9LA7HX4HB"  # Replace with your actual API key
    url = "https://www.alphavantage.co/query"
    params = {
        "function": "TIME_SERIES_INTRADAY",  # API function to get intraday time series
        "symbol": symbol,  # Stock symbol
        "interval": "1min",  # Interval for the data
        "apikey": api_key,
    }

    response = requests.get(url, params=params)
    data = response.json()  # Parse the response as JSON

    # Check for errors in the response
    if "Error Message" in data:
        return f"Error fetching data: {data['Error Message']}"

    # Extract the latest time series data
    time_series = data.get("Time Series (1min)", {})
    if not time_series:
        return "No data available for the given symbol."

    # Get the latest time data (most recent available entry)
    latest_time = next(iter(time_series))
    latest_data = time_series[latest_time]

    # Create a formatted string with the stock details
    output = (
        f"Stock: {symbol}\n"
        f"Time: {latest_time}\n"
        f"Open: {latest_data['1. open']}\n"
        f"High: {latest_data['2. high']}\n"
        f"Low: {latest_data['3. low']}\n"
        f"Close: {latest_data['4. close']}\n"
        f"Volume: {latest_data['5. volume']}\n"
    )

    return output

# Example usage
stock_symbol = "AAPL"  # Replace with the desired stock symbol
if api_req == 'yes':
    live_data = fetch_live_data(stock_symbol)
    print(live_data)
else:
    live_data = None
```
