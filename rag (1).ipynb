# **Data Transformation**

```python
from langchain_community.document_loaders.csv_loader import CSVLoader
from langchain_core.documents import Document

loader = CSVLoader(file_path='/content/aapl_stock_data.csv', encoding='utf-8', csv_args={'delimiter': ','})
data = loader.load()

def chunk_data(data, chunk_size=50):
    """Splits data into chunks of specified row size."""
    return [data[i:i+chunk_size] for i in range(0, len(data), chunk_size)]

# Split the data into chunks of 50 rows each
chunked_data = chunk_data(data, chunk_size=100)

# Each chunk can now be treated as a separate document
documents = [Document(page_content=str(chunk)) for chunk in chunked_data]

print(documents[0].page_content)
```

# **Retrieval and Chain Creation**

```python
class MyClass:
    def setup_qa_chain(self, public_data_db):
        try:
            public_retriever = public_data_db.as_retriever(search_type="similarity",
                                                        search_kwargs={"k":3})
            # public_retriever = public_data_db.as_retriever()
            # public_retriever = MultiQueryRetriever.from_llm(
            #                        retriever=public_data_db.as_retriever(),
            #                        llm=llm )
            # docs = public_retriever.get_relevant_documents({'question' : "Retrieve the opening price for 2023-02-01"})
            docs = public_retriever.invoke("What was the highest price last week?")
            print(docs)

            system_prompt = ("""
            You are an assistant with access to historical trading data for a specific stock or cryptocurrency (eg.,AAPL).
            The dataset includes fields such as price (opening, highest, lowest, closing), volume, and timestamps.

            Inputs:
              'query': {question}
              'Context': {context}

            Your task is to answer user queries strictly based on this dataset.

            Instructions:
            - For questions about specific dates, return the exact value from the dataset (e.g., opening price, highest price, or volume).
            - For questions about a time range, calculate and return the relevant value based on the specified period.
            - If no data is available for a requested date or period, respond with: "Data not available for the requested period."
            - Format your response as a complete, grammatically correct sentence, making it sound like a conversation.
            - Always be concise and factual, without adding any extra information or speculations.

            Respond only with the relevant information while ensuring clarity and a conversational tone and ensure it was a accurate information.
            """)


            # prompt = ChatPromptTemplate.from_template(template)
            prompt = ChatPromptTemplate.from_template(system_prompt)
            output_parser = StrOutputParser()


            # chain = RetrievalQA.from_chain_type(
            #     llm,
            #     chain_type="stuff",
            #     retriever=public_retriever.get_relevant_documents({question : "what is your name"}),
            #     return_source_documents=True,
            #     chain_type_kwargs = {"prompt": template}
            # )
            # query = "How many AI publications in 2021?"
            # result = chain({"query": query})
            # print(result)
            chain = RunnableMap({
                            "context": lambda x: public_retriever.get_relevant_documents(x["question"]),
                            "question": lambda x: x["question"]
                          }) | prompt | llm | output_parser

            return chain

        except Exception as e:
            return str(e)

# Usage
# Check if index folder is empty

#     # Select embeddings to be used
# embeddings = HuggingFaceEmbeddings(
#         model_name=modelPath,     # Provide the pre-trained model's path
#         model_kwargs=model_kwargs, # Pass the model configuration options
#         encode_kwargs=encode_kwargs # Pass the encoding options
#     )

llm = ChatGoogleGenerativeAI(model="gemini-pro", google_api_key=os.environ['GOOGLE_API_KEY'])

embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001")


persist_directory = "/content/chromaDB"
public_data_db = Chroma.from_documents(documents = documents,
    persist_directory=persist_directory,
                            embedding=embeddings)

# Load Chromadb from disk
public_data_db.persist()
public_data_db = Chroma(
    persist_directory=persist_directory,
                            embedding_function=embeddings)


# print(public_data_db.get())

print("QA chain setup in progress..")

instance = MyClass()  # Create an instance of the class
return_val = instance.setup_qa_chain(public_data_db)

if type(return_val) != str:
        print("SUCCESS: Chatbot is setup")
        chatbot = return_val
        print(chatbot)
else:
        print("ERROR: Chatbot is not setup")
        print("The error is : " + return_val)
```
